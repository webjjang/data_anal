```python
# 머신 러닝할 데이터 준비
import gdown
gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)

# 다운 받은 파일을 pandas에 세팅하기
import pandas as pd

# low_memory=False 설정에 따라 데이터가 달라 질 수 있다.
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
from sklearn.model_selection import train_test_split

# 훈련용 데이터와 테스트 데이터로 분할하기 - 랜덤하게 선택
train_set, test_set = train_test_split(ns_book7, random_state=42)
```

```python
print(len(train_set),len(test_set))
```

```python
# 선형회귀 모델 -> 2차 배열
x_train = train_set[["도서권수"]]
y_train = train_set["대출건수"]

print(x_train.shape, y_train.shape)
```

```python
# 머신러닝 훈련시키기
from sklearn.linear_model import LinearRegression

lr = LinearRegression() # 선행 회귀 모델 객체 생성
lr.fit(x_train, y_train) # 학습 데이터로 모델 학습 진행
```

```python
# 훈련 모델 평가하기
x_test = test_set[["도서권수"]]
y_test = test_set["대출건수"]

lr.score(x_test, y_test)
```

```python
lr.fit(y_train.to_frame(), y_train)
lr.score(y_test.to_frame(), y_test)
```

```python
print(lr.coef_, lr.intercept_)
```

```python
# 로지스틱 회귀 모델 - 이진 분류
# 대출 권수의 평균
borrow_mean = ns_book7["대출건수"].mean()
y_train_c = y_train > borrow_mean
y_test_c = y_test > borrow_mean
```

```python
# 로지스틱 라이브러리 import
from sklearn.linear_model import LogisticRegression

logr = LogisticRegression() # 객체 생성
logr.fit(x_train, y_train_c) # 훈련
logr.score(x_test, y_test_c) # 평가
```
