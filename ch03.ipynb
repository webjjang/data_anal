```python
# 필요한 파일 다운 받기
import gdown
gdown.download("https://bit.ly/3RhoNho", "ns_202104.csv", quiet=False)
```

```python
# pandas로 데이터 일어오기
import pandas as pd
ns_df = pd.read_csv("ns_202104.csv", low_memory=False)
ns_df.head()
```

```python
ns_book = ns_df.loc[:, "번호":"등록일자"]
ns_book.head()
```

```python
# 항목명 출력
print(ns_df.columns)
print(type(ns_df.columns))
# 인덱스 번호 0번 항목 이름 출력
print(ns_df.columns[0])
```

```python
# Unnamed: 13 항목 이름이 아닌 항목에 대해서 True 출력하자.
ns_df.columns != "Unnamed: 13"
# True인 항목의 데이터만 선택하자.
selected_columns = ns_df.columns != "Unnamed: 13"
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
selected_columns = ns_df.columns != "부가기호"
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
# drop 메서드를 이용한 데이터 제거
# 항목 한개 제거
# axis=0 : 행을 제거, axis=1 : 열을 제거
ns_book = ns_df.drop("Unnamed: 13", axis=1)
ns_book.head()
# 항목 여러개 제거
# 원본 데이터는 변경되지 않는다.
print(ns_df.columns)
ns_book = ns_df.drop(["부가기호","Unnamed: 13"], axis=1)
ns_book.head()
```

```python
# inplace=True : 원본 데이터 변경해라. 원래 drop은 원본 데이터 변경하지 않는다.
ns_book.drop("주제분류번호", axis=1, inplace=True)
ns_book.head()
```

```python
# NaN - Not A Number
#ns_book = ns_df.dropna(axis=1, how="all") # how="all" -항목의 모든 데이터가 NaN일 때 처리
ns_book = ns_df.dropna(axis=1) # 한개라도 NaN이면 처리
ns_book.head()
```
