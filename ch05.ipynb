```python
import gdown
gdown.download("https://bit.ly/3pK7iuu","ns_book7.csv",quiet=False)
```

```python
import pandas as pd
ns_book7 = pd.read_csv("ns_book7.csv", low_memory=False)
ns_book7.head()
```

```python
# scatter 그래프 그리기
import matplotlib.pyplot as plt
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# figure : 도표의 크기 세팅, figsize=(x축의 크기,y축의 크기) : 단위 인치입니다.
plt.figure(figsize=(9,6))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# 기본 그래프의 크기 확인
print(plt.rcParams['figure.figsize'])
# 기본 해상도 확인
print(plt.rcParams['figure.dpi'])
```

```python
# figure : 도표의 크기 세팅, figsize=(x축의 크기,y축의 크기) : 단위 인치입니다.
plt.figure(figsize=(900/72,600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# 코랩의 설정값을 바꾼다. - 적용된다. : 출력할 때 경계선에 붙었는 것은 출력범위를 벗어나서 출력이 안된다. 출력 여백 추가해서 이미지 만든다.
%config InlineBackend.print_figure_kwargs = {"bbox_inches" : None}
plt.figure(figsize=(900/72,600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# rcParams = runtime configue params - 해상도 100으로 바꾸기
plt.rcParams["figure.dpi"] = 100
```

```python
# 산점도 마커 모양 바꾸기
# o -> *
# 마커 모양 확인
# plt.rcParams["scatter.marker"]
# 마커 모양 변경 - o, *, +, x 등으로 변경
plt.rcParams["scatter.marker"] = "x"
```

```python
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# 여러개의 그래프 그리기 - 서브 플롯
# fig, axs = plt.subplots(2)
# fig, axs = plt.subplots(2, figsize=(6,8))
fig, axs = plt.subplots(1, 2, figsize=(10,4))

# 첫번째 그래프 설정
axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
# 첫번째 그래프 제목 설정
axs[0].set_title("Scatter plot")
# x축 이름 설정 - 도서 권수
axs[0].set_xlabel("number of books")
# y축 이름 설정 - 대출건수
axs[0].set_ylabel("borrow count")

# 두번째 그래프 설정
axs[1].hist(ns_book7['대출건수'], bins = 100)
# 작은 데이터 표시를 위해 로그 처리
axs[1].set_yscale('log')
# 첫번째 그래프 제목 설정
axs[1].set_title("Histrogram")
# x축 이름 설정 - 대출 건수
axs[1].set_xlabel("borrow count")
# y축 이름 설정 - 빈도수 - 한글을 사용하려면 font 설치해서 사용해야한다.
axs[1].set_ylabel("frequency")

plt.show()
# fig.show()
```
